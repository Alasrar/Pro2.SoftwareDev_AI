```python
#Building Models
models_list = []
models_list.append(('CART', DecisionTreeClassifier()))
models_list.append(('SVM', SVC()))
models_list.append(('NB', GaussianNB()))
models_list.append(('KNN', KNeighborsClassifier()))
```

```python
#Cross_Validation
num_folds = 10
results = []
names = []
param=[]
for name, model in models_list:
    kfold = KFold(n_splits=num_folds)#, random_state=123)
    start = time.time()
    cv_results = cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')
    p=model.get_params()
    param.append(model.get_params())
    end = time.time()
    results.append(cv_results)
    names.append(name)
    print(f'model: {name}, params: {p}, name: {model}\n')
    print( "%s: %f (%f) (run time: %f)" % (name, cv_results.mean(), cv_results.std(), end-start))
    print('\n\n')
```

```python
fig = plt.figure(figsize=(6,4))
fig.suptitle('Performance Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

```python
import warnings
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score,cross_validate
num_folds = 10
results1 = []
names1 = []
param1=[]
scoring = {
    'accuracy': make_scorer(accuracy_score),
    'precision': make_scorer(precision_score, average='weighted'),
    'recall': make_scorer(recall_score, average='weighted'),
    'f1_score': make_scorer(f1_score, average='macro')
}
pipelines = []
pipelines.append(('ScaledCART', Pipeline([('Scaler', StandardScaler()),('CART',DecisionTreeClassifier())])))
pipelines.append(('ScaledSVM', Pipeline([('Scaler', StandardScaler()),('SVM', SVC( ))])))
pipelines.append(('ScaledNB', Pipeline([('Scaler', StandardScaler()),('NB',GaussianNB())])))
pipelines.append(('ScaledKNN', Pipeline([('Scaler', StandardScaler()),('KNN',KNeighborsClassifier())])))
results1 = []
names1 = []
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    kfold = KFold(n_splits=num_folds)#, random_state=123)
    for name1, model in pipelines:
        start1 = time.time()
        # cv_results = cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')
        cv_results1 = cross_validate(model, X_train, Y_train, cv=kfold, scoring=scoring)#,'precision','recall','f1_macro'])

        end1 = time.time()
        results1.append(cv_results1)
        names1.append(name1)
        # print( "%s: (run time: %f)" % (name, end-start))
print(names1)

for i in results1:
    # print(f"test_accuracy:{np.mean(i['test_accuracy'])}")#\ntest_precision:{np.meann(i['test_precision'])}\ntest_recall:{np.mean(i['test_recall'])}\ntest_f1_score:{np.mean(i['test_f1_score'])}')
    print(f"Accuracy:{np.mean(i['test_accuracy']):.3f}\nPrecision:{np.mean(i['test_precision']):.3f}\nRecall:{np.mean(i['test_recall']):.3f}\nF1_score:{np.mean(i['test_f1_score']):.3f}")
    print('\n')
```

```python
#Scaling means transforing Data[btween 0,1]
scaler = StandardScaler().fit(X_train)
rescaledX = scaler.transform(X_train)
#Tunning on SVM between the result of it is too high
c_values = [0.1, 0.3, 0.5, 0.7, 0.9, 1.0, 1.3, 1.5, 1.7, 2.0]
#ploy, linear, rbf, sigmid = ttpes of model kernal
kernel_values = ['linear', 'poly', 'rbf', 'sigmoid']
param_grid = dict(C=c_values, kernel=kernel_values)
model = SVC()
kfold = KFold(n_splits=num_folds)#, random_state=21)
grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring='accuracy', cv=kfold)
grid_result = grid.fit(rescaledX, Y_train)
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```
